```julia
using Distributions
```

```julia
function num(epsilon, delta, binnum)
    chi2 = Chisq(binnum - 1)
    return ceil(quantile(chi2, 1.0 - delta)/(2*epsilon))
end
```

```julia
fig, (axl, axr) = plt.subplots(ncols=2, figsize=(10,4))
bs = Vector(2:10)
n = [num(0.1, 0.01, b) for b in bs]
axl.set_title("bin:2-10")
axl.plot(bs,n)

bs = Vector(2:100000)
n = [num(0.1, 0.01, b) for b in bs]
axr.set_title("bin:2-100000")
axr.plot(bs,n)
```
