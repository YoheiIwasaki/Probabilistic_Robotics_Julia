```julia
mutable struct Robot <: AbstractRobot
    pose
    r
    color
    agent
    poses
    sensor
end

#TODO: search inheritance
Robot(pose) = Robot(pose, 0.2, "black", nothing, [pose], nothing)
Robot(pose, color) = Robot(pose, 0.2, color, nothing, [], nothing)
Robot(pose, agent) = Robot(pose, 0.2, "black", agent, [pose], nothing)
Robot(pose, agent, color) = Robot(pose, 0.2, color, agent, [pose], nothing)
Robot(pose, agent, sensor, color) = Robot(pose, 0.2, color, agent, [pose], sensor)
```

```julia
world = World(10, 1.0)
for i in 1:100
    circling = Agent(0.2, 10.0/180*pi)
    r = Robot([0.0,0.0,0.0], circling, nothing, "black")
    append(world, r)
end
draw(world)
```
